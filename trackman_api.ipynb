```python
import requests
import pandas as pd
import os

# Step 1: Authenticate and Get Access Token
auth_url = "https://login.trackmanbaseball.com/connect/token"
auth_data = {
    "client_id": "UnivofCaliforniaSanDiego-Analysis",
    "client_secret": "7eb4bec7-35a0-4ea5-b26e-d38f4231d98c",
    "grant_type": "client_credentials"
}

auth_response = requests.post(auth_url, data=auth_data)
if auth_response.status_code == 200:
    access_token = auth_response.json().get("access_token")
    print("✅ Access Token received.")
else:
    print("❌ Authentication Error:", auth_response.status_code, auth_response.text)
    exit()

# Step 2: Get Video Metadata for a Specific Session
video_metadata_url = "https://dataapi.trackmanbaseball.com/api/v1/media/game/videotokens/230a6d85-9972-4b45-ada8-4cd41c8847be"
headers = {
    "Authorization": f"Bearer {access_token}",
    "accept": "application/json"
}

response = requests.get(video_metadata_url, headers=headers)

# Step 3: Parse response and extract video links
if response.status_code == 200:
    data = response.json()
    
    if isinstance(data, list):
        df = pd.DataFrame(data)
        if 'utcDate' in df.columns:
            df['date'] = pd.to_datetime(df['utcDate']).dt.date
        
        print("🟢 Video Metadata Retrieved")
        
        # Save metadata to CSV
        df.to_csv("trackman_api/game_sessions_jan_media.csv", index=False)

        # Step 4: Download or Stream videos (Assuming "url" or similar key exists)
        os.makedirs("videos", exist_ok=True)
        for idx, row in df.iterrows():
            video_url = row.get("videoUrl") or row.get("url")  # Replace with correct key
            if video_url:
                video_filename = f"videos/video_{idx}.mp4"
                try:
                    print(f"⬇️ Downloading video {idx} from {video_url}")
                    video_resp = requests.get(video_url, stream=True)
                    with open(video_filename, "wb") as f:
                        for chunk in video_resp.iter_content(chunk_size=1024*1024):
                            f.write(chunk)
                    print(f"✅ Saved: {video_filename}")
                except Exception as e:
                    print(f"❌ Failed to download video {idx}: {e}")
            else:
                print(f"⚠️ No video URL found for index {idx}")
    else:
        print("❌ Unexpected response format:", data)
else:
    print("❌ Error getting metadata:", response.status_code, response.text)
```
